```python
# импорт модуля
import Graph
```

```python
# Определяем наш граф и пишем, какое количество человек должно быть в графе
graph = Graph.Graph(5)

# Добавление узла с именем человека
graph.add_vertex("Bob")
graph.add_vertex("Dan")
graph.add_vertex("Alice")
graph.add_vertex("Kourt")
graph.add_vertex("Amy")

# Добавление связи между именами
graph.add_edge("Bob", "Amy")
graph.add_edge("Alice", "Amy")
graph.add_edge("Dan", "Kourt")
graph.add_edge("Dan", "Amy")
graph.add_edge("Dan", "Bob")
graph.add_edge("Alice", "Kourt")

# Функция дя вывода матрицы смежности
graph.adjacency_matrix()
```

```python
# Функция для вывода списка смежности
print(graph.adjacency_list())
```

```python
# Функция дя проверки, является ли граф деревом
print(graph.is_tree())
```

```python
# Функция для отрисовки матрицы смежности
print(graph.drow_adjacency_graph())
```

```python
# Функция для отрисовки самого графа
print(graph.drow_graph())
```
